```python
import qrcode
from cryptography.fernet import Fernet

# 🔹 암호화 키 생성 (한 번 생성 후 저장해서 사용)
key = Fernet.generate_key()
cipher = Fernet(key)

# 🔹 QR 코드에 넣을 데이터 (이름은 암호화)
name = "홍길동"
encrypted_name = cipher.encrypt(name.encode()).decode()

# 🔹 QR 코드에 넣을 URL (이름은 숨기고 URL만 표시)
base_url = "https://cunning-bear-2x0aoa-dev-ed.trailblaze.my.site.com/coffeetest/s"
qr_data = f"{base_url}?data={encrypted_name}"  # URL만 보이고 이름은 암호화

# 🔹 QR 코드 생성
qr = qrcode.make(qr_data)
qr.save("qrcode_hidden_name.png")
qr
```
